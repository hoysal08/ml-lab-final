```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("../pima (1).csv")
data.head()
```

```python
X=data.drop("Outcome",axis=1)
Y=data.Outcome
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,stratify=Y,test_size=0.1,random_state=34)
```

```python
from sklearn.ensemble import RandomForestClassifier

instance=RandomForestClassifier()
instance.fit(X_train,Y_train)
y_prec=instance.predict(X_test)
accuracy_score(Y_test,y_prec)
```

```python
print(classification_report(Y_test,y_prec))
```

```python
from sklearn.tree import DecisionTreeClassifier
instance=DecisionTreeClassifier()
instance.fit(X_train,Y_train)
y_precc=instance.predict(X_test)
accuracy_score(Y_test,y_precc)
```

```python
import graphviz
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(instance)
```
